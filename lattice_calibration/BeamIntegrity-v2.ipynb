```python
# Constants 
x1 = 9.04e-3
x2 = 1.84e-4


def CheckBeamIntegrity( b ):
    def mWEr_f( r, wI, wR):
        return x1 * wI * wR  / np.sqrt(r)
    def mWkHz2D_f( r, wI, wR):
        return x2 / ( 1./(wI**4) + r/(wR**4) )
    def mWkHz2L_f( r, wI, wR):
        return x2 / ( 1./(wI**4) + r/(wR**4) + np.sqrt(r)/wI/(wR**3) + np.sqrt(r)/wR/(wI**3) )
    
    
    def errL( x ):
        r=x[0]
        wI=x[1]
        wR=x[2]
        err1 = np.abs( mWEr_f( r, wI, wR) - b['mWEr'] )  / b['mWEr']
        err2 = np.abs( mWkHz2D_f( r, wI, wR) - b['mW_kHz2_Dimple'] ) / b['mW_kHz2_Dimple']
        err3 = np.abs( mWkHz2L_f( r, wI, wR) - b['mW_kHz2_Lattice'] ) / b['mW_kHz2_Lattice']
        err4 = np.abs( wI-wR )
        return err1**2. + err2**2.  + err3**2.
    
    def errL_same_waist( x ):
        r=x[0]
        wI=x[1]
        wR=x[2]
        err1 = np.abs( mWEr_f( r, wI, wR) - b['mWEr'] )  / b['mWEr']
        err2 = np.abs( mWkHz2D_f( r, wI, wR) - b['mW_kHz2_Dimple'] ) / b['mW_kHz2_Dimple']
        err3 = np.abs( mWkHz2L_f( r, wI, wR) - b['mW_kHz2_Lattice'] ) / b['mW_kHz2_Lattice']
        err4 = np.abs( wI-wR )
        return err1**2. + err2**2.  + err3**2. + err4**2.
    
    
    # define reasonable ranges to explore the parameters
    r_  = np.linspace( 0.4, 1.2, 14 )
    wI_ = np.linspace( 34., 56., 36 ) 
    wR_ = np.linspace( 34., 56., 36 )
    

    
    table = [] 
    
    for r in r_:
        for wR in wR_:     
            wIs = []
            for wI in wI_:
                error = errL( [r, wI, wR] ) 
                wIs.append( [r,wI,wR, mWEr_f(r,wI,wR), mWkHz2D_f(r,wI,wR), mWkHz2L_f(r,wI,wR), error ])
            wIs = np.array( wIs )
            wIs = wIs[ np.argsort( wIs[:,6] ) ] 
            table.append( wIs[0] ) 
            
            
    table = np.array(table)            
                
                
    fig2 = plt.figure(figsize=(17,8.)) 
    gs2 = matplotlib.gridspec.GridSpec( 2,3, hspace=0.35, wspace=0.4) 
    axA = fig2.add_subplot( gs2[0,0] )
    axB = fig2.add_subplot( gs2[0,1] )
    axC = fig2.add_subplot( gs2[0,2] ) 
    axD = fig2.add_subplot( gs2[1,0] ) 
    axE = fig2.add_subplot( gs2[1,1] ) 
    
    
    for ax in [axA, axB, axC, axD, axE ]:
        ax.xaxis.set_major_locator( matplotlib.ticker.MaxNLocator(6))
        ax.yaxis.set_major_locator( matplotlib.ticker.MaxNLocator(6))
    
    for ax in [axA, axB, axC, axD, axE ]:
        ax.set_xlabel('$r$')
        ax.set_ylabel('$w_{RETRO}$')
        ax.set_xlim( r_.min(), r_.max() )
        ax.set_ylim( wR_.min(), wR_.max() )
    
    def meshplot( ax, i, j, k , **kwargs):
        zero = kwargs.pop('zero',0.)
        contours = kwargs.pop('contours',None)
        contours_ratio = kwargs.pop('contours_ratio',None)
        x = table[:,i]
        y = table[:,j]
        z = table[:,k] - zero
        xi = np.linspace( x.min(), x.max(), 300)
        yi = np.linspace( y.min(), y.max(), 300)
        zq = matplotlib.mlab.griddata(x, y, z, xi,yi)
        im0 =ax.pcolormesh( xi, yi, zq , cmap = plt.get_cmap('rainbow'), **kwargs)
        plt.axes( ax)
        plt.colorbar(im0) 
        
        if contours is not None:
            c0 = ax.contour(xi, yi, zq, contours, linewidths = 0.5, colors = 'k')
            plt.clabel(c0, inline=1, fontsize=10)
            
        if contours_ratio is not None:
            zq = matplotlib.mlab.griddata(x, y, z/y, xi,yi)
            c0 = ax.contour(xi, yi, zq, contours_ratio, linewidths = 0.5, colors = 'k')
            plt.clabel(c0, inline=1, fontsize=10)
    
    axA.set_title('$w_{IN}$') 
    meshplot( axA, 0,2,1, contours_ratio=[0.95,1.0,1.05] )
    
    
    zero = b['mWEr']
    err  = b['mWEr_err']
    axB.set_title('mW/Er (%.2f +/- %.2f)'%(zero,err) )
    meshplot( axB, 0,2,3 , zero=zero, contours=[-err,0,err]) 
    
    zero = b['mW_kHz2_Dimple']
    err = b['mW_kHz2_Dimple_err']
    axC.set_title('mW/kHz^2 Dimple  (%.2f +/- %.2f)'%(zero,err))
    meshplot( axC, 0,2,4 , zero=zero, contours=[-err,0,err]) 
    
    zero = b['mW_kHz2_Lattice']
    err = b['mW_kHz2_Lattice_err']
    axD.set_title('mW/kHz^2 Lattice (%.2f +/- %.2f)'%(zero,err))
    meshplot( axD, 0,2,5 , zero=zero, contours=[-err,0,err]) 
    
    axE.set_title('error')
    meshplot( axE, 0,2,6 , vmin=0., vmax=0.02) 
    
    
    from scipy.optimize import minimize

    x0 = [0.8, 44., 44.]
    res = minimize( errL, x0, method='Nelder-Mead', options={'disp': True})
    print "Minimization of the error function:"
    print "\n\t r = %.3g" % res.x[0] 
    print "\twI = %.3g" % res.x[1] 
    print "\twR = %.3g" % res.x[2] 
    
    print "\n"
    res = minimize( errL_same_waist, x0, method='Nelder-Mead', options={'disp': True})
    print "\nMinimization of the error function,"
    print "with constraint wI=wR:"
    print "\n\t r = %.3g" % res.x[0] 
    print "\twI = %.3g" % res.x[1] 
    print "\twR = %.3g" % res.x[2] 
    
    ukb_over_mpi2 = 8.55e3  
    print "\n"
    print "With the results for wI=wR we obtain the"
    print "following slopes for dimple and lattice"
    print "\n"
    print "DIMPLE  = %.2f"%( 1./ ( ukb_over_mpi2 * ( 2 + 2*res.x[0] ) / (np.pi * res.x[1]**4 ) ) )
    print "LATTICE = %.2f"%( 1./ ( ukb_over_mpi2 * ( 2 + 2*res.x[0] + 4*np.sqrt( res.x[0] ) ) / (np.pi * res.x[1]**4 ) ) )
    



```
