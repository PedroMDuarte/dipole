```python
# Constants 
x1 = 9.04e-3
x2 = 1.84e-4


def CheckBeamIntegrity( b ):
    def mWEr_f( r, wI, wR):
        return x1 * wI * wR  / np.sqrt(r)
    def nuD_f( r, wI, wR):
        return np.sqrt( b['PD'] / x2 * ( 1./(wI**4) + r/(wR**4) ) ) 
    def nuL_f( r, wI, wR):
        return np.sqrt( b['PL'] / x2 * ( 1./(wI**4) + r/(wR**4) ) ) 
    
    def errL( x ):
        r=x[0]
        wI=x[1]
        wR=x[2]
        err1 = np.abs( mWEr_f( r, wI, wR) - b['mWEr'] )
        err2 = np.abs( nuL_f( r, wI, wR) - b['nuL'] )
        err3 = np.abs( wI-wR )
        return (err1/b['mWEr'])**2 + (err2/b['nuL'])**2 
    
    def errL_same_waist( x ):
        r=x[0]
        wI=x[1]
        wR=wI
        err1 = np.abs( mWEr_f( r, wI, wR) - b['mWEr'] )
        err2 = np.abs( nuL_f( r, wI, wR) - b['nuL'] )
        err3 = np.abs( wI-wR )
        return (err1/b['mWEr'])**2 + (err2/b['nuL'])**2 
    
    # define reasonable ranges to explore the parameters
    r_  = np.linspace( 0.4, 1.2, 14 )
    wI_ = np.linspace( 34., 56., 36 ) 
    wR_ = np.linspace( 34., 56., 36 )
    

    
    table = [] 
    
    for r in r_:
        for wR in wR_:     
            wIs = []
            for wI in wI_:
                error = errL( [r, wI, wR] ) 
                wIs.append( [r,wI,wR, mWEr_f(r,wI,wR), nuL_f(r,wI,wR), error ])
            wIs = np.array( wIs )
            wIs = wIs[ np.argsort( wIs[:,5] ) ] 
            table.append( wIs[0] ) 
            
            
    table = np.array(table)            
                
                
    fig2 = plt.figure(figsize=(17,3.5)) 
    gs2 = matplotlib.gridspec.GridSpec( 1,4, hspace=0.65 ) 
    axA = fig2.add_subplot( gs2[0,0] )
    axB = fig2.add_subplot( gs2[0,1] )
    axC = fig2.add_subplot( gs2[0,2] ) 
    axD = fig2.add_subplot( gs2[0,3] ) 
    
    
    for ax in [axA, axB, axC, axD ]:
        ax.xaxis.set_major_locator( matplotlib.ticker.MaxNLocator(6))
        ax.yaxis.set_major_locator( matplotlib.ticker.MaxNLocator(6))
    
    for ax in [axA, axB, axC, axD ]:
        ax.set_xlabel('$r$')
        ax.set_ylabel('$w_{RETRO}$')
    
    def meshplot( ax, i, j, k , **kwargs):
        x = table[:,i]
        y = table[:,j]
        z = table[:,k] 
        xi = np.linspace( x.min(), x.max(), 300)
        yi = np.linspace( y.min(), y.max(), 300)
        zq = matplotlib.mlab.griddata(x, y, z, xi,yi)
        im0 =ax.pcolormesh( xi, yi, zq , cmap = plt.get_cmap('rainbow'), **kwargs)
        plt.axes( ax)
        plt.colorbar(im0) 
    
    axA.set_title('$w_{IN}$') 
    meshplot( axA, 0,2,1 )
    
    axB.set_title('mW/Er')
    meshplot( axB, 0,2,3 ) 
    
    axC.set_title('nu_L')
    meshplot( axC, 0,2,4 ) 
    
    axD.set_title('error')
    meshplot( axD, 0,2,5 , vmin=0., vmax=0.02) 
    
    
    from scipy.optimize import minimize

    x0 = [0.8, 44., 44.]
    res = minimize( errL, x0, method='Nelder-Mead')
    print "Minimization of the mWEr and nu_L error function:"
    print "\n\t r = %.3g" % res.x[0] 
    print "\twI = %.3g" % res.x[1] 
    print "\twR = %.3g" % res.x[2] 
    
    res = minimize( errL_same_waist, x0, method='Nelder-Mead')
    print "\nMinimization of the mWEr and nu_L error function,"
    print "with constraint wI=wR:"
    print "\n\t r = %.3g" % res.x[0] 
    print "\twI = %.3g" % res.x[1] 
    print "\twR = %.3g" % res.x[2] 



```
